```python
import os

from dotenv import load_dotenv
from supabase import create_client, Client

# 1. Load Environment Variables
load_dotenv()
```

```python
url: str = os.environ.get("SUPABASE_URL")
key: str = os.environ.get("SUPABASE_KEY")
anon_key : str = os.environ.get("SUPABASE_ANON_KEY")

supabase: Client = create_client(url, key)
anon_supabase = create_client(url, anon_key)
```

```python
# Add secret

secret_dict={
    'secret':"mysecret111",
    'secret_name':"heyheymimi",
    'description':"testing"
}
response = supabase.rpc('vault_add_secret', params=secret_dict).execute()
print("secret uuid:",response.data)
```

```python
# Read secret - unexisting secret

response = supabase.rpc('vault_get_secret', params={'secret_name':"asd"}).execute()
print(response.data)

print('---')
# existing one

response = supabase.rpc('vault_get_secret', params={'secret_name':"heyheymimi"}).execute()
print(response.data)
```

```python
# Update secret - existing one

secret_dict={
    'uuid':"11d8ee07-fd44-439f-a3fc-90165f7832e9",
    'secret':"mysecret222",
    'secret_name':"heyheymimi",
}
response = supabase.rpc('vault_update_secret', params=secret_dict).execute()
print(response.data)

print('---')
# - unexisting one

secret_dict={
    'uuid':"11d8ee832e9",
    'secret':"mysecret222",
    'secret_name':"heyheymimi",
}
response = supabase.rpc('vault_update_secret', params=secret_dict).execute()
print(response.data)
```

```python
# Testing with insufficient role

secret_dict={
    'secret':"mysecret111",
    'secret_name':"heyheymimi",
    'description':"testing"
}
response = anon_supabase.rpc('vault_add_secret', params=secret_dict).execute()
print("secret uuid:",response.data)
```
